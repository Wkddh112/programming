데이터사용량과 월평균 요금 비교,
데이터사용량과 초과요금 비교,
를 통해서 월평균과 초과요금이 데이터 사용량이랑 얼마나 비례하는지 알아보겠습니다.

```python
plt.title("데이터사용량의 월평균요금", fontsize=15)
plt.scatter(telco["DataAmount"], telco["MonthlyFee"], color='r', alpha=.5)
plt.xlabel("월평균요금", fontsize=13)
plt.ylabel("데이터사용량", fontsize=13) 
plt.grid()
plt.show()
```

```python
plt.title("데이터사용량의 초과요금", fontsize=15)
plt.scatter(telco["DataAmount"], telco["OverFee"], color='r', alpha=.5)
plt.xlabel("데이터사용량", fontsize=13)
plt.ylabel("초과요금", fontsize=13) 
plt.grid()
plt.show()
```

```python
plt.title("월평균요금과 초과요금", fontsize=15)
plt.scatter(telco["MonthlyFee"], telco["OverFee"], color='r', alpha=.5)
plt.xlabel("월평균요금", fontsize=13)
plt.ylabel("초과요금", fontsize=13) 
plt.grid()
plt.show()
```

그래프에서 보는것과 같이 데이터사용량의 월평균요금제는 많이 비례하는것이 보여지고, 데이터사용량의 초과요금은 월평균 요금제 보다는 덜 비례 되는것을 볼수있다.

```python
#"DataAmount". "MonthlyFee". 두 피처간의 상관계수를 계산합니다.
corr = telco[["DataAmount", "MonthlyFee"]].corr(method = 'pearson')
print(corr)
```

```python
#DataAmount와 상관계수가 가장 맞는것은 "MonthlyFee"로 나타난다.
tels = ['DataAmount','OverFee','MonthlyFee','Callcenter','CalMins']
corr = telco[tels].corr(method = 'pearson')
print(corr)
```

```python
type(corr)
```

```python
import seaborn as sns

#corr 행렬 히트맵을 시각화합니다.
tels = ['DataAmount','OverFee','MonthlyFee','Callcenter','CalMins']
sns.set(font_scale = 1)
hm = sns.heatmap(corr.values,
                cbar=True,
                annot=True,
                square=True,
                fmt='.2f',
                yticklabels=tels,
                xticklabels=tels)

plt.tight_layout()
plt.show()
```

이로 봤을때, 색이 밝을 수록 연관도가 높고 어두울수록 연관도가 낮다는것을 알수있습니다.
