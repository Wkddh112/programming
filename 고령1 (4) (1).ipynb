```python
from konlpy.tag import Hannanum
from collections import Counter
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

para = open('C:/Users/dudql/Documents/1/국내고령관광.txt', encoding='UTF-8').read()

tag = Hannanum()
nouns = tag.nouns(para)
nouns = [n for n in nouns if len(n) > 1]

count = Counter(nouns)
tags = count.most_common(50)

icon = Image.open('picture.png').convert("RGBA")
mask = Image.new("RGB", icon.size, (255,255,255))
mask.paste(icon,icon)
mask = np.array(mask)

my_path='C:/Users/dudql/AppData/Local/Microsoft/Windows/Fonts/BMHANNAAir_ttf.ttf'

wordcloud = WordCloud(font_path=my_path, 
                      background_color='white', 
                      width=1200, 
                      height=800,
                     mask=mask).generate_from_frequencies(dict(tags))

fig = plt.figure(figsize=(15,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```
