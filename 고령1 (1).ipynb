```python
from collections import Counter
from konlpy.tag import Komoran
tag = Komoran()

f = open("C:/Users/dudql/Documents/1/국내고령관광.txt", 'r', encoding = "UTF-8")
data = f.read()
f.close()
print(data[:100])

nouns = tag.nouns(data)
nouns = [n for n in nouns if len(n) > 1]

count = Counter(nouns)

tags = count.most_common(10)
tags
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


my_path='C:/Windows/Fonts/Bahnschrift.ttf'

wordcloud = WordCloud(font_path=my_path, 
                      background_color='white', 
                      width=1200, 
                      height=800).generate_from_frequencies(dict(count))

fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 텍스트 파일 열기, 트럼프대통령 취임사
text = open('C:/Users/dudql/Documents/1/국내고령관광.txt', encoding='UTF-8').read()

# 워드 클라우드 이미지 생성하기
wordcloud = WordCloud(background_color='white', 
                      width=1920, 
                      height=1080).generate(text)

# 화면에 출력하기
fig = plt.figure()
plt.imshow(wordcloud, interpolation='bilinear', cmap='YlOrBr')
plt.axis('off')
```

```python
import numpy as np
from PIL import Image
icon = Image.open('C:/Users/dudql/Documents/1/노인.jpg').convert("RGBA")
mask = Image.new("RGB", icon.size, (255,255,255))
mask.paste(icon,icon)
mask = np.array(mask)

wordcloud = WordCloud(background_color='white', 
                      max_font_size = 100,
                      max_words=1000, 
                      mask=mask).generate(text)

# 화면에 출력하기
fig = plt.figure(figsize=(15,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```
